## 1.1 Create a two-dimensional vector and plot it on a graph

```python
import matplotlib.pyplot as plt
import numpy as np

# Two Dimensional Vector
blue = [.5, .5]

# Plot the Vector
plt.arrow(0,0, blue[0], blue[1],head_width=.05, head_length=0.05, color ='blue')
plt.xlim(0,1)          
plt.ylim(0,1)
plt.title("2D Vector")
plt.show()
```

## 1.2 Create a three-dimensional vector and plot it on a graph

```python
from mpl_toolkits.mplot3d import Axes3D
import numpy as np

vector = np.array([[0, 0, 0, .5, .5, .5]])

X, Y, Z, U, V, W = zip(*vector)
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.quiver(X, Y, Z, U, V, W, length=1)
ax.set_xlim([0, 1])
ax.set_ylim([0, 1])
ax.set_zlim([0, 1])
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
plt.show()
```

## 1.3 Scale the vectors you created in 1.1 by $5$, $\pi$, and $-e$ and plot all four vectors (original + 3 scaled vectors) on a graph. What do you notice about these vectors? 

```python
import math

# Original Vector
blue = [.5, .5]

# Get negative e
e = math.e
negative_e = -1*e

# Scale the vector
green = np.multiply(5, blue)
red = np.multiply(math.pi, blue)
orange = np.multiply(negative_e, blue)

# Plot the scaled vectors
plt.arrow(0,0, green[0], green[1],head_width=.05, head_length=0.05, color ='green')
plt.arrow(0,0, red[0], red[1],head_width=.05, head_length=0.05, color ='red')
plt.arrow(0,0, blue[0], blue[1],head_width=.05, head_length=0.05, color ='blue')
plt.arrow(0,0, orange[0], orange[1],head_width=.05, head_length=0.05, color ='orange')
plt.xlim(-2,3)          
plt.ylim(-2,3)
plt.title("Scaled Vectors")
```

## 1.8 Find $||g||$ and then find $||h||$. Which is longer?

\begin{align}
\vec{g} = \begin{bmatrix} 1 \\ 1 \\ 1 \\ 8 \end{bmatrix}
\qquad
\vec{h} = \begin{bmatrix} 3 \\3 \\ 3 \\ 3 \end{bmatrix}
\end{align}

```python
# Column Vectors
g = np.array([[1],[1],[1],[8]])
h = np.array([[3],[3],[3],[3]])

norm_g = np.linalg.norm(g)
norm_h = np.linalg.norm(h)

print("Norm of g:", norm_g)
print("Norm of h:", norm_h)
       
# g is longer than h.

print('\n')

# Row Vectors
g = np.array([1,1,1,8])
h = np.array([3,3,3,3])

norm_g = np.linalg.norm(g)
norm_h = np.linalg.norm(h)

print("Norm of g:", norm_g)
print("Norm of h:", norm_h)
       
# g is longer than h.
```

# Part 3 - Square Matrices

## 3.1 Find $IG$ (be sure to show your work) 😃

\begin{align}
G= 
\begin{bmatrix}
12 & 11 \\
7 & 10 
\end{bmatrix}
\end{align}

```python
# Create Identity Matrix
I = np.array([[1,0],[0,1]])

# Create G Matrix
G = np.array([[12,11],[7,10]])

#Multiplying G and Identity Matrix should give back G matrix
print("I*G = \n", np.matmul(I,G))
```

## 3.2 Find $|H|$ and then find $|J|$.

\begin{align}
H= 
\begin{bmatrix}
12 & 11 \\
7 & 10 
\end{bmatrix}
\qquad
J= 
\begin{bmatrix}
0 & 1 & 2 \\
7 & 10 & 4 \\
3 & 2 & 0
\end{bmatrix}
\end{align}


```python
H = np.array([[12,11],[7,10]])
J = np.array([[0,1,2],[7,10,4],[3,2,0]])
             
print("Determinant of H:", np.linalg.det(H))
print("Determinant of J:", np.linalg.det(J))
```

## 3.3 Find $H^{-1}$ and then find $J^{-1}$

```python
H_inverse = np.linalg.inv(H)
J_inverse = np.linalg.inv(J)

print("Inverse of H: \n", H_inverse)
print("\n")
print("Inverse of J: \n", J_inverse)
```

## 3.4 Find $HH^{-1}$ and then find $G^{-1}G$. Is $HH^{-1} == G^{-1}G$? Why or Why not?

```python
G_inverse = np.linalg.inv(G)

H_H_inverse = np.matmul(H,H_inverse)

G_inverse_G = np.matmul(G,G_inverse)

print("H*H^{-1} \n", H_H_inverse)
print("\n")
print("G*G^{-1} \n", G_inverse_G)
print("\n")
print("Are the two matrices equal?:", np.array_equal(H_H_inverse, G_inverse_G))
```
