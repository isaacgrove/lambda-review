```python
import pandas as pd

# Load via URL - pass url as string to pd.read_csv()

flag_data_url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/flags/flag.data'
# If you curl it, .data actually looks like a csv

df = pd.read_csv(flag_data_url)

df.head()


# Load from file (Colab)
#
# from google.colab import files
# uploaded = files.upload()
# df = pd.read_csv('flag (1).data', names=column_headers)
```

```python
# Sometimes you have to fix headers yourself
column_headers = ['name', 'landmass', 'zone', 'area', 'population', 'language', 
                  'religion', 'bars', 'stripes', 'colours', 'red', 'green', 
                  'blue', 'gold', 'white', 'black', 'orange', 'mainhue', 
                  'circles', 'crosses', 'saltires', 'quarters', 'sunstars', 
                  'crescent', 'triangle', 'icon', 'animate', 'text', 'topleft', 
                  'botright']
df = pd.read_csv(flag_data_url, names=column_headers)
df.head()


# Or you can make a header row with no names
#
# df = pd.read_csv(flag_data_url, header=None)
```

```python
# Read in another dataset, look at descriptive statistics, check for nulls, make some plots

column_headers = ['age', 'workclass', 'fnlwgt', 'education', 'education-num', 
                 'marital-status', 'occupation', 'relationship', 'race', 'sex', 
                 'capital-gain', 'capital-loss', 'hours-per-week', 
                 'native-country', 'income']

adult_data = 'https://archive.ics.uci.edu/ml/machine-learning-databases/adult/adult.data'
df = pd.read_csv(adult_data, names=column_headers)
df.head()
```

```python
df.describe()


# df.describe(exclude='number')
# df.describe(include='all')
```

```python
df.dtypes
```

```python
df.isnull().sum()
```

```python
# Upon further inspection, null values are encoded as " ?"

import numpy as np
df = df.replace({" ?": np.NaN})

df.isnull().sum()
```

```python
df['native-country'][14]
```

```python
# Visualizations

df['education-num'].value_counts().sort_index().plot();
```
