6. What is Shapley Values?
```
--it's a game theory concept. If players cooperate in a positive sum game and create a surplus, how do you distribute the surplus?
--how important was each player, and what payoff can they expect? The Shapley value provides one possible answer to this question.
--it works on an individual prediction basis, how much each feature impacts that prediction
```

7. How do you get ROC AUC score??
```
--in a binary classifier, you set a threshold for how "easy" it is to classify something as True. You then plot True positive rate against False positive rate
--i.e. if you have a really high threshold of 1, everything gets classified as Postive. You get all the true positives, but incorrectly classify all the negatives as positive as well.
--Area under the curve in the chart is how well you did. You want to capture all the true positives without accidentally capturing false positives.
--0 to 1, 1 is best.
```
