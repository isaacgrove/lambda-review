```python
# Rename columns

df.rename(columns = {'Start':'start', 
                     'End':'end', 
                     'Sleep quality': 'sleep_quality_dirty',
                     'Time in bed': 'time_in_bed', 
                     'Wake up': 'wake_up_mood', 
                     'Sleep Notes': 'sleep_notes', 
                     'Heart rate': 'heart_rate', 
                     'Activity (steps)': 'step_count'}, inplace = True)
```

```python
# Convert sleep quality to float

# I could also divide by 100 here to normalize sleep_quality

def sleep_quality_to_float(cell_contents):
  return float(cell_contents.strip().strip('%'))

new_sleep_quality_list = []

for cell in df['sleep_quality_dirty']:
  new_sleep_quality_list.append(sleep_quality_to_float(cell))

df['sleep_quality'] = pd.Series(new_sleep_quality_list)
```

```python
# Drop redundant column and columns that are just missing values

df = df.drop(['wake_up_mood', 'sleep_notes', 'heart_rate', 'sleep_quality_dirty'], axis=1)
```

```python
# Convert start and end to datetime objects

df['start'] = pd.to_datetime(df['start'], infer_datetime_format=True)
df['end'] = pd.to_datetime(df['end'], infer_datetime_format=True)
```

```python
# Convert time_in_bed to float and call it "hours_in_bed"

df['hours_in_bed'] = df['time_in_bed'].str.split(':').apply(lambda x: float(x[0]) + float(x[1])/60)
```

```python
# Engineer a few features

# A rudimentary exhaustion metric. Steps I took in the day divided by hours of sleep
df['steps_per_hour_slept'] = df['step_count'] / df['hours_in_bed']

# The hour of the day (in military time) I went to bed, and the hour I got up.
df['start_hour'] = df['start'].dt.hour
df['end_hour'] = df['end'].dt.hour
# Make these categorical for encoding.
df['start_hour'] = df['start_hour'].astype(str)
df['end_hour'] = df['end_hour'].astype(str)
```

```python
# Remove outliers (lowest 0.5% sleep quality)
df = df[df['sleep_quality'] >= np.percentile(df['sleep_quality'], 0.5)]

# I do this because a couple of the lowest quality sleeps were actually naps
# and don't really apply to what I want to learn from the data.
```

```python
# First, split data into train and test sets

from sklearn.model_selection import train_test_split

train, test = train_test_split(df, train_size=0.80, test_size=0.20, 
                               random_state=42)

train.shape, test.shape
```

```python
# Mean baseline for regression. Report MAE

# Train
guess = train['sleep_quality'].mean()
errors = guess - train['sleep_quality']
mean_absolute_error = errors.abs().mean()
print(f'Mean: {guess:.2f}% \n')
print(f'Mean baseline train MAE: {mean_absolute_error:.2f}% \n')
#(f'Train MAE: {mae:.2f}%')
# Test
guess = test['sleep_quality'].mean()
errors = guess - test['sleep_quality']
mean_absolute_error = errors.abs().mean()
print(f'Mean baseline test MAE: {mean_absolute_error:.2f}%')
```

```python
# Set up X features matrices and y target vectors for first model

features = ['hours_in_bed']
target = ['sleep_quality']

X_train = train[features]
y_train = train[target]
X_test = test[features]
y_test = test[target]
```

```python
# Linear regression with 1 feature: hours_in_bed
#
# Beats mean baseline: 8.75% --> 7.95% 
#
# Model note: I didn't impute means because there are none to impute

from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

model = LinearRegression()
model.fit(X_train, y_train)



# Train
y_pred = model.predict(X_train)
mae = mean_absolute_error(y_train, y_pred)
mse = mean_squared_error(y_train, y_pred)
rmse = np.sqrt(mse)
r2 = r2_score(y_train, y_pred)
print(f'Linear Regression train MAE: {mae:.2f}% \n')
#print(f'Train R2: {r2:.2f} \n')

# Test
y_pred = model.predict(X_test)
mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
r2 = r2_score(y_test, y_pred)
print(f'Linear Regression test MAE: {mae:.2f}%')
#print(f'Test R2: {r2:.2f}')
```

```python
# Model 2: multiple regression (4 features, one-hot encoding)

# Slightly overfits compared to 1-feature

import category_encoders as ce
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

features = ['hours_in_bed', 'steps_per_hour_slept', 'start_hour', 'end_hour']
target = ['sleep_quality']

X_train = train[features]
y_train = train[target]
X_test = test[features]
y_test = test[target]

# One Hot Encode categoricals
encoder = ce.OneHotEncoder(use_cat_names=True)
X_train2 = encoder.fit_transform(X_train)
X_test2 = encoder.transform(X_test)


model2 = LinearRegression()

model2.fit(X_train2, y_train)

y_pred = model2.predict(X_train2)
mae = mean_absolute_error(y_train, y_pred)
print(f'Multiple regression train MAE: {mae:.2f}%\n')

y_pred = model2.predict(X_test2)
mae = mean_absolute_error(y_test, y_pred)
print(f'Multiple regression test MAE: {mae:,.2f}% \n')

# "Get coefficients." Encoding means there's a lot of them
beta0 = model2.intercept_
beta1 = model2.coef_
print(f'Equation of model2 line: y = {beta0} + {beta1}x1')
```

```python
# Get permutation importances for Multiple Regression Model

import eli5
from eli5.sklearn import PermutationImportance

permuter = PermutationImportance(
    model2, 
    scoring='neg_mean_absolute_error', 
    n_iter=5, 
    random_state=42
)

permuter.fit(X_train2, y_train)
```

```python
# Display permutation importances

feature_names = X_train2.columns.tolist()

eli5.show_weights(
    permuter, 
    top=None, # No limit: show permutation importances for all features
    feature_names=feature_names
 ) # must be a list
```

```python
# SelectKBest and permutation importances are BOTH ways to select # of features
```

```python
# SelectKBest linear regression features

# change k in "selector" to see which features get selected for the model in the next cell

from sklearn.feature_selection import SelectKBest, f_regression

selector = SelectKBest(score_func=f_regression, k=1)

# The .values.ravel() was a Stack Overflow fix for an error we didn't have in class
# regarding y column vectors versus 1d arrays.
X_train_selected = selector.fit_transform(X_train2, y_train.values.ravel())
X_test_selected = selector.transform(X_test2)


selected_mask = selector.get_support()
all_names = X_train2.columns
selected_names = all_names[selected_mask]
unselected_names = all_names[~selected_mask]

print('Features selected:')
for name in selected_names:
    print(name)

print('\n')
print('Features not selected:')
for name in unselected_names:
    print(name)
```

```python
# Display K Best linear regression models

# 1 feature is actually best; using more features just overfits the model.
# The max features test is the same as model2 above

for k in range(1, len(X_train2.columns)+1):
    print(f'{k} features')
    
    selector = SelectKBest(score_func=f_regression, k=k)
    X_train_selected = selector.fit_transform(X_train2, y_train.values.ravel())
    X_test_selected = selector.transform(X_test2)

    model = LinearRegression()
    model.fit(X_train_selected, y_train)
    y_pred = model.predict(X_train_selected)
    mae = mean_absolute_error(y_train, y_pred)
    print(f'Train MAE: {mae:,.2f}%')

    y_pred = model.predict(X_test_selected)
    mae = mean_absolute_error(y_test, y_pred)
    print(f'Test MAE: {mae:,.2f}% \n')
```

```python
# Random Forest

from sklearn.ensemble import RandomForestRegressor
from sklearn.impute import SimpleImputer
from sklearn.pipeline import make_pipeline

features = ['hours_in_bed', 'steps_per_hour_slept', 'start_hour', 'end_hour']
target = ['sleep_quality']

X_train = train[features]
y_train = train[target]
X_test = test[features]
y_test = test[target]

forest_pipeline = make_pipeline(
    ce.OrdinalEncoder(), 
    SimpleImputer(strategy='median'), 
    RandomForestRegressor(random_state=0, n_jobs=-1)
)

# Fit on train, score on val
forest_pipeline.fit(X_train, y_train)

# Train
y_pred = forest_pipeline.predict(X_train)
mae = mean_absolute_error(y_train, y_pred)
print(f'Train MAE: {mae:,.2f}%')

# Test
y_pred = forest_pipeline.predict(X_test)
mae = mean_absolute_error(y_test, y_pred)
print(f'Test MAE: {mae:,.2f}%')
```
